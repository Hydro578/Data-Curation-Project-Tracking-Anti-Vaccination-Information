```python
houston = pd.read_csv("houston.csv")
```

```python
houston = houston[["date","confirmed"]]
```

```python
start_date = '2021-12-20'
end_date = '2022-02-07'
dates = (houston['date'] > start_date) & (houston['date'] <= end_date) #narrow downs the dates we want for LA

houston= houston.loc[dates]
ax1 = houston.plot(
        title = 'Confirmed Total Cases Late Dec to Mid Feb',
        x = 'date',
        y = 'confirmed',
        legend = None,
        figsize = (20,10)
    )
ax1.set_ylabel("Confirmed Cases (in millions)")
```
